```python
# Mencetak angka genap dari 1 hingga 100
print("Angka Genap:")
for i in range(2, 101, 2):
    print(i)

# Mencetak angka ganjil dari 1 hingga 100
print("Angka Ganjil:")
for i in range(1, 101, 2):
    print(i)
```

Nomor 2

```python
# Menghitung hasil penjumlahan dari 0 hingga 100
total_penjumlahan = 0

for number in range(101):
    print(total_penjumlahan ,"+", number)
    total_penjumlahan += number
    print(total_penjumlahan)

print("Hasil penjumlahan dari 1 hingga 100 adalah:", total_penjumlahan)
```


## Bagian 2

Nomor 1

```python
n = 1000
total = (n * (n + 1) * (2 * n + 1)) / 6
print("Jumlah 1000 bilangan kuadrat pertama adalah:", total)
```

Nomor 2

```python
total = 0
for number in range(1, 2500):
    if number % 10 == 0 or number % 7 == 0:
        total += number
print("Hasil penjumlahan bilangan asli kelipatan 10 atau 7 di bawah 2500 adalah:", total)
```
